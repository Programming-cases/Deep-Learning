```python
import torch as th
import matplotlib.pyplot as plt
```

## Derivada simple:  

### Variable puntual:
**Primer ejemplo:**

$\mathit{f(x)=x²}$<br>$\mathit{f\,'(x)= \frac{d f}{d x} = 2x} $

No te olvides de agregar el valor de **`2`** como **`2.0`**: 


```python
#Definiendo el valor de x=2:
x= th.tensor(2.0,requires_grad=True)
#Definiendo a la función y= x²:
y= pow(x,2)
#Derivando la función y:
y.backward()
#Evaluando a la función derivada:
x.grad
```

**Segundo ejemplo:**<br>
$\mathit{f(x)=x²+2x+1}$<br>$\mathit{f\,'(x)= \frac{d f}{d x} = 2x+2} $

```python
#x=4:
x= th.tensor(4.0, requires_grad=True)
#z= x²+2x+1:
z= pow(x,2)+2*x+1
#z':
z.backward()
#z'(x=4):
x.grad
```

## Derivada parcial:
$\mathit{f(u,v)=uv+u²}$<br>$\mathit{\frac{d f}{d u} = v+2u} $ <br> $\mathit{\frac{d f}{d v} = u} $

```python
u= th.tensor(1.0,requires_grad=True)
v= th.tensor(2.0,requires_grad=True)
f=u*v+pow(u,2)
f.backward()
```

```python
#Evaluando v:
v.grad
```

```python
#Evaluando u:
u.grad
```

### Con variables discretas:
$\mathcal{x:-10,-9,...,9,10}$<br>  $f(x)= x²$<br> $f'(x)=2x$

```python
x=th.linspace(-10,10,10,requires_grad=True)
Y=pow(x,2)
y=th.sum(Y)#¡¡IMPORTANTE A AÑADIR EN DOMINIO DE VARIABLES DISCRETAS!!
y.backward()
```

```python
plt.style.use('seaborn')
plt.figure(figsize=(16,6))
#No hace falta llamar a Numpy:
plt.plot(x.detach().numpy(),Y.detach().numpy(),label='función',c='steelblue')
plt.plot(x.detach().numpy(),x.grad.detach().numpy(),label='derivada',c='cyan')
plt.legend()
```
