# Analizing the Logistic function  <br> 
<br> $$\large \sigma(z)=\frac{1}{1+e^{-z}}$$

```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('ggplot')
```

Setting the domain:

```python
X= np.arange(-5,5,0.1)
```

Creating the Logisctic function: 

```python
def Lfunction(w,b):
    z= X*w+b
    LF=pow((1+np.exp(-1*z)),-1)
    return LF
```

## 1. Analizing w with b fixed

```python
b=1
plt.figure(figsize=(17,7))
plt.plot(X,Lfunction(1,0),color='blue',label='Main',linewidth=3)
for i in range(-2,3):
    plt.plot(X,Lfunction(i,b),label= 'z={0}x+{1}'.format(i,b))
plt.legend(loc=8)
plt.title('Logistic funcion with b={}'.format(b))
plt.show()
```

- When w is negative, the function is reversed with respect to x axis
- It's not worth applying the function if **z is constant**

## 2. Analizing b with w fixed

```python
w=-3
plt.figure(figsize=(17,7))
plt.plot(X,Lfunction(1,0),color='blue',label='Main',linewidth=3)
for i in range(-2,3):
    plt.plot(X,Lfunction(w,i),label= 'z={0}x+{1}'.format(w,i))
plt.legend(loc=8)
plt.title('Logistic funcion with w={}'.format(w))
plt.show()
```
